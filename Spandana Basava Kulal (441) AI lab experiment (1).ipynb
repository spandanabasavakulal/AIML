```python
# Experiment 3
import heapq
def uniform_cost_search(graph, start, goal):
    priority_queue = [(0, start, [])]
    visited = set()
    while priority_queue:
        cost, node, path = heapq.heappop(priority_queue)
        if node in visited:
            continue
        path = path + [node]
        visited.add(node)
        if node == goal:
            return cost, path
        for neighbor, edge_cost in graph.get(node, []):
            if neighbor not in visited:
                heapq.heappush(priority_queue, (cost + edge_cost, neighbor, path))
    return float("inf"), []
graph = {
    'A': [('B', 4), ('C', 3)],
    'B': [('F', 5), ('E', 12)],
    'C': [('E', 10), ('D', 7)],
    'D': [('E', 2)],
    'E': [('Z', 5)],
    'F': [('Z', 16)]
}
start_node = 'A'
goal_node = 'Z'
cost, path = uniform_cost_search(graph, start_node, goal_node)
print(f"UCS path from {start_node} to {goal_node} is: {path}")
print(f"Total cost: {cost}")
```

```python
# Experiment 4
import heapq
class Node:
    def __init__(self, name, g=0, h=0):
        self.name = name
        self.g = g
        self.h = h
        self.f = g + h
        self.parent = None
    def __lt__(self, other):
        return self.f < other.f
def a_star_search(graph, start, goal, heuristic):
    open_list = []
    heapq.heappush(open_list, Node(start, 0, heuristic[start]))
    closed_list = set()
    while open_list:
        current_node = heapq.heappop(open_list)
        if current_node.name == goal:
            path = []
            total_cost = current_node.g
            while current_node:
                path.append(current_node.name)
                current_node = current_node.parent
            return path[::-1], total_cost

        closed_list.add(current_node.name)
        for neighbor, cost in graph[current_node.name].items():
            if neighbor in closed_list:
                continue
            g_cost = current_node.g + cost
            h_cost = heuristic[neighbor]
            neighbor_node = Node(neighbor, g_cost, h_cost)
            neighbor_node.parent = current_node
            heapq.heappush(open_list, neighbor_node)
    return None
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 1},
    'D': {'B': 2, 'G': 2},
    'E': {'B': 5},
    'F': {'C': 1, 'G': 3},
    'G': {'D': 2, 'F': 3}
}
heuristic = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 3,
    'E': 2,
    'F': 2,
    'G': 0
}
start_node = 'A'
goal_node = 'G'
path, total_cost = a_star_search(graph, start_node, goal_node, heuristic)
print(f"Least-cost path from {start_node} to {goal_node}: {path}")
print(f"Total cost: {total_cost}")
```

```python
# Experiment 5
import random
def hill_climbing (objective_function, initial_solution, step_size = 0.1, max_iterations = 1000):
    current_solution = initial_solution
    current_value = objective_fucntion(current_solution)
    for _ in range (max_iterations):
        neighbor = current_solution + random.uniform (-step_size, step_size)
        neighbor_value = objective_function (neighbor)
        if neighbor_value > current_value:
            current_solution = neighbor
            current_value = neighbor_value
    return current_solution, current_value
def objective_fucntion (x):
    return -x**2 + 4*x
initial_solution = random.uniform (-10,10)
best_solution, best_value = hill_climbing (objective_fucntion, initial_solution)
print(f"Best solution : {best_solution}")
print(f"Best value: {best_value}")
```

```python
# Experiment 6
MAX = 1000
MIN = -1000
def minmax(depth, nodeIndex, maximizingPlayer, values, alpha, beta):
    if depth == 3:
        return values[nodeIndex]
    if maximizingPlayer:
        best = MIN
        for i in range(0, 2):
            val = minmax(depth + 1, nodeIndex * 2 + i, False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        best = MAX
        for i in range(0, 2):
            val = minmax(depth + 1, nodeIndex * 2 + i, True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)
            if beta <= alpha:
                break
        return best
if __name__ == "__main__":
    values = [3, 5, 6, 9, 1, 2, 0, 1]                                                                                                                                                                                                                                                                                                           
    print("The optimal value is:", minmax(0, 0, True, values, MIN, MAX))
```

```python
# Experiment 8
# Facts
parents = {
    'mary': ['john', 'susan'],
    'david': ['john', 'susan'],
    'linda': ['mary'],
    'steve': ['mary']
}
# Gender information
male = {'john', 'david', 'steve'}
female = {'susan', 'mary', 'linda'}
# Rules
def parent(x, y):
    return x in parents.get(y, [])
def father(x, y):
    return parent(x, y) and x in male
def mother(x, y):
    return parent(x, y) and x in female
def grandparent(x, y):
    return any(parent(x, z) and parent(z, y) for z in parents)
def sibling(x, y):
    if x == y:
        return False
    x_parents = set(parents.get(x, []))
    y_parents = set(parents.get(y, []))
    return bool(x_parents & y_parents)
# Sample Queries
print("father(john, mary):", father("john", "mary"))         # True
print("mother(susan, david):", mother("susan", "david"))     # True
print("grandparent(john, linda):", grandparent("john", "linda")) # True
print("sibling(mary, david):", sibling("mary", "david"))     # True
```

```python
# Symptoms for each person
symptoms = {
    'john': {'fever', 'cough', 'shortness_of_breath'},
    'alice': {'headache', 'runny_nose', 'sneezing'}
}
# Disease definitions
diseases = {
    'covid': {'fever', 'cough', 'shortness_of_breath'},
    'flu': {'fever', 'cough', 'runny_nose'},
    'common_cold': {'runny_nose', 'sneezing', 'headache'}
}
# Diagnosis function
def diagnose(person):
    person_symptoms = symptoms.get(person, set())
    diagnosis = []
    for disease, required_symptoms in diseases.items():
        if required_symptoms.issubset(person_symptoms):
            diagnosis.append(disease)
    return diagnosis
# Example usage
john_diagnosis = diagnose("john")
print(f"Disease(s) diagnosed for John: {john_diagnosis}")
```
